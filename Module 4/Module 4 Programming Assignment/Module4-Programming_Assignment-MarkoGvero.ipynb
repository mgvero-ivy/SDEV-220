```python
#16.4

import sqlite3

conn = sqlite3.connect('books.db')
cursor = conn.cursor()

cursor.execute('''CREATE TABLE IF NOT EXISTS books (title TEXT, author TEXT, year INTEGER)''')

conn.commit()
conn.close()
```

```python
#16.5

import csv
import sqlite3

conn = sqlite3.connect('books.db')
cursor = conn.cursor()

with open("books2.csv", newline='') as file:
    reader = csv.reader(file)
    next(reader)  # Skip header row
    for row in reader:
        cursor.execute("INSERT INTO books (title, author, year) VALUES (?, ?, ?)", row)

conn.commit()
conn.close()
```

```python
#16.8

from sqlalchemy import create_engine, text

engine = create_engine('sqlite:///books.db') #sets engine to connect to the database

connection = engine.connect()

result = connection.execute(text("SELECT title FROM books ORDER BY title")) #sets result to the titles of the books in the database, ordered alphabetically

for row in result:
    print(row[0])

connection.close()
```
