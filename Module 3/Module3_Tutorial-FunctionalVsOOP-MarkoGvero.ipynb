```python
#|Part 1

class Solution:
    def sort012(self, arr):
        # code here
        n = len(arr)
        lo = 0
        hi = n - 1
        mid = 0

        while mid <= hi:
            if arr[mid] == 0: #if mid is 0 it is swapped with lo and both lo and mid are incremented by 1
                arr[lo], arr[mid] = arr [mid], arr[lo]
                lo += 1
                mid += 1
            elif arr[mid] == 1: #if mid is 1 then mid is incremented by 1
                mid += 1
            else: #if mid is neither 0 nor 1 then it must be 2 and mid is swapped with hi and hi is decremented by 1
                arr[mid], arr[hi] = arr[hi], arr[mid]
                hi -= 1
        return(arr)
```

```python
#|test cases for part 1
arr1 = [0, 1, 2, 0, 1, 2]
arr2 = [0, 1, 1, 0, 1, 2, 1, 2, 0, 0, 0, 1]

sol = Solution()
print(sol.sort012(arr1))
print(sol.sort012(arr2))
```

```python
#|Part 2
class Solution2:
    def firstSearch(self, arr, k):
        # Code Here


        n = len(arr)
        lo = 0
        hi = n - 1
        answer = -1

        while lo <= hi: #checks to see if lo is greater than hi so it stops if no match is found
            guess = (lo + hi) // 2

            if arr[guess] == k: #checks the guess against the desired number
                answer = guess  #saves a variable of the location of a found match
                hi = guess - 1  #decrements hi to continue searching left for a lower match
            elif arr[guess] < k:
                lo = guess + 1  #increments lo to search further right in the array
            else:
                hi = guess - 1  #decrements hi to search further left in the array

        return answer
```

```python
#|test cases for part 2
arr1 = [1, 2, 3, 4, 5]
arr2 = [11, 22, 33, 44, 55]
arr3 = [1, 1, 1, 1, 2]

sol = Solution2()
print(sol.firstSearch(arr1, 4))
print(sol.firstSearch(arr2, 445))
print(sol.firstSearch(arr3, 1))
```
