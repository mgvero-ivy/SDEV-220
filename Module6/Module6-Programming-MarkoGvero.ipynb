15.1 Use multiprocessing to create three separate processes. Make each one wait a random number of seconds between zero and one, print the current time, and then exit.

```python
# 13.1
from datetime import date

today = date.today()

today_string_formatted = today.strftime("%Y-%m-%d")

with open("today.txt", "w") as file:
    file.write(today_string_formatted)

print("Date written to today.txt", today_string_formatted)
```

```python
# 13.2
with open("today.txt", "r") as file:
    today_string = file.read()

print("read from file: " , today_string)
```

```python
# 13.3
parsed_date = date.strptime(today_string.strip(), "%Y-%m-%d")

print("parsed date: " , parsed_date)
print("type: " , type(parsed_date))
print("year: ", parsed_date.year)
print("month: ", parsed_date.month)
print("day: " , parsed_date.day)
```

```python
# 15.1
from multiprocessing import Process
from datetime import datetime
import time
import random

def wait_random():
    wait_time = random.random()
    time.sleep(wait_time)

    c_time = datetime.now()

    c_time_formatted = c_time.strftime("%H:%M:%S")

    print(wait_time , "seconds waited, the current time is ", c_time_formatted)


#wait_random()

if __name__ == "__main__":
    p1 = Process(target=wait_random)
    p2 = Process(target=wait_random)
    p3 = Process(target=wait_random)

    p1.start()
    p2.start()
    p3.start()

    p1.join()
    p2.join()
    p3.join()

    print("p1 exit code:", p1.exitcode)
    print("p2 exit code:", p2.exitcode)
    print("p3 exit code:", p3.exitcode)

    print("multiprocessing completed")
```
